Feature Engineering <br>

Add Column = HP_high_low

```python
hp_mean = df["HP"].mean()

def set_hp(val):
    if val < hp_mean:
        return "HP Low"
    elif val == hp_mean:
        return "HP neutral"
    else:
        return "HP High"
```

```python
df["HP_high_low"] = df["HP"].apply(set_hp)
df
```

Add Column = Speed_high_low

```python
speed_mean = df["Speed"].mean()

def set_speed(val):
    if val < speed_mean:
        return "Speed Low"
    elif val == speed_mean:
        return "Speed neutral"
    else:
        return "Speed High"
```

```python
df["Speed_high_low"] = df["HP"].apply(set_hp)
df
```
